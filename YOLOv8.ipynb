```python
from ultralytics import YOLO

model = YOLO("yolov8n.yaml")  

model.info()  
model.train(data="/content/drive/MyDrive/yolofinal/data.yaml", epochs=100)  
```

```python
model.predict("/content/OIP (1).jpeg")
```
